```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
import os
os.path.exists(r'C:\Users\MUBARAKALI R\Desktop\Nila Project.csv')
```

```python
dataset=pd.read_csv(r'C:\Users\MUBARAKALI R\Desktop\Nila Project.csv')
dataset.head()
```

```python
dataset.info()
```

```python
# 1. Convert 'TotalCharges' to Numeric
dataset['TotalCharges']=dataset['TotalCharges'].replace(" ",'0')
dataset['TotalCharges']=dataset['TotalCharges'].replace('float')
```

```python
dataset.describe ()
```

```python
ax = sns.countplot(x='Churn', data=dataset)  # use exact name
ax.bar_label(ax.containers[0])
plt.show()
```

```python
sns.boxplot(x='Churn', y='MonthlyCharges', data=dataset)
plt.title('Monthly Charges vs Churn')
plt.show()
```

```python
corr = dataset.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
